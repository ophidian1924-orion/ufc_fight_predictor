```python
import pandas as pd
df = pd.read_csv("all_fights.csv")
```

```python
df.duplicated().sum()
```

```python
#Convert fight_date to datetime and sequence it chronologically
df["fight_date"] = pd.to_datetime(df["fight_date"])
df = df.sort_values('fight_date').reset_index(drop=True)
```

```python
#Check null values
df.isnull().sum().sort_values(ascending=False)
```

```python
#Change categorical data into binary values
df['red_stance'] = df['red_stance'].map({'Southpaw': 1, 'Orthodox': 0})
df['blue_stance'] = df['blue_stance'].map({'Southpaw': 1, 'Orthodox': 0})
df['gender'] = df['gender'].map({'MALE': 1, 'FEMALE': 0})
```

```python
df['red_stance'] = df['red_stance'].fillna(2)
df['blue_stance'] = df['blue_stance'].fillna(2)
df['red_stance'].isna().sum()
```

```python
df['weight_class'].isna().sum()
```

```python
#weight_class has too many categories to change into binary manually. Use OneHotEncoder.
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(df[['weight_class']])
encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(['weight_class']),
    index=df.index
)
encoded_df.head()
```

```python
encoded_df.isna().sum()
```

```python
df1 = df.copy()
df1.head()
```

```python
#Join encoded df with main df
df2 = df1.join(encoded_df, how='inner')
df2.isna().sum()
```

```python
df2.info()
```

```python
df2['title_bout'] = df2['title_bout'].map({'t': 1, 'f': 0})
df2['red_winner'] = df2['red_winner'].map({'t': 1, 'f': 0})
df2.head()
```

```python
df2 = df2.drop(columns=['red_return', 'blue_return'])
```

```python
df2.info()
```

```python
#Turn American odds into probability
#Start Feature Engineering
def american_to_prob(odds):
    if odds < 0:
        return abs(odds)/(abs(odds) + 100)
    else:
        return 100/(abs(odds) + 100)


```

```python
df2["red_implied_prob"] = df2["red_odds"].apply(american_to_prob)

df2["blue_implied_prob"] = df2["blue_odds"].apply(american_to_prob)
```

```python
df2['prob_diff'] = (df2['red_implied_prob'] - df2['blue_implied_prob'])
```

```python
df2.head()
```

```python
#Win Percentage
df2['red_win_pct'] = df2['red_wins']/(df2['red_wins']+df2['red_losses'])
df2['red_win_pct'].info()
```

```python
df2['blue_win_pct'] = df2['blue_wins']/(df2['blue_wins']+ df2['blue_losses'])
df2['win_pct_diff'] = df2['red_win_pct'] - df2['blue_win_pct']
```

```python
#KO Rate
df2['red_ko_rate'] = df2['red_wins_by_ko']/df2['red_wins']
df2['blue_ko_rate'] = df2['blue_wins_by_ko']/df2['blue_wins']
```

```python
#Submission Rate
df2['red_sub_rate'] = df2['red_wins_by_submission']/df2['red_wins']
df2['blue_sub_rate'] = df2['blue_wins_by_submission']/df2['blue_wins']
```

```python
#Finish Rate
df2['red_finish_rate'] = (df2['red_wins_by_ko']+ df2['red_wins_by_submission'])/df2['red_wins']
df2['blue_finish_rate'] = (df2['blue_wins_by_ko'] + df2['blue_wins_by_submission'])/df2['blue_wins']
```

```python
#Total Fights
df2['red_experience'] = df2['red_wins'] + df2['red_losses']
df2['blue_experience'] = df2['blue_wins'] + df2['blue_losses']
```

```python
#Average Round per Fight
df2['red_round_per_fight'] = df2['red_total_rounds_fought']/df2['red_experience']
df2['blue_round_per_fight'] = df2['blue_total_rounds_fought']/df2['blue_experience']
```

```python
#Reach Per Height
df2['red_reach_per_height'] = df2['red_reach_cms']/df2['red_height_cms']
df2['blue_reach_per_height'] = df2['blue_reach_cms']/df2['blue_height_cms']
```

```python
#Number of days since the fighter fought last
last_fight_date = {}
df2['red_days_since_last'] = None
df2['blue_days_since_last'] = None

for index, row in df2.iterrows():

    red = row["red_fighter"]
    blue = row["blue_fighter"]

    fight_date = row["fight_date"]

    # Red fighter
    if red in last_fight_date:
        days = (fight_date - last_fight_date[red]).days
        df2.loc[index, "red_days_since_last"] = days
    else:
        df2.loc[index, "red_days_since_last"] = None

    # Blue fighter
    if blue in last_fight_date:
        days = (fight_date - last_fight_date[blue]).days
        df2.loc[index, "blue_days_since_last"] = days
    else:
        df2.loc[index, "blue_days_since_last"] = None

    
    last_fight_date[red] = fight_date
    last_fight_date[blue] = fight_date
```

```python
df2.tail()
```

```python
#Debut Fighters
df2["red_debut"] = df2["red_days_since_last"].isna().astype(int)
df2["blue_debut"] = df2["blue_days_since_last"].isna().astype(int)
```

```python
df2["activity_diff"] = (
    df2["red_days_since_last"] -
    df2["blue_days_since_last"]
)
```

```python
df2.info()
```

```python
#
df2 = df2.drop(columns=['red_fighter', 'blue_fighter', 'red_wins', 'red_losses', 'blue_wins', 'blue_losses', 'red_wins_by_ko',
                       'blue_wins_by_ko', 'weight_class', 'fight_date', 'red_wins_by_submission', 'blue_wins_by_submission'])
```

```python
df2['red_days_since_last'] = pd.to_numeric(df2['red_days_since_last'], errors='coerce')
df2['blue_days_since_last'] = pd.to_numeric(df2['blue_days_since_last'], errors='coerce')
df2['activity_diff'] = pd.to_numeric(df2['activity_diff'], errors='coerce')
df2.info()
```

```python
df2.to_csv(
    "processed_ufc_fights.csv",
    index=False
)
```
